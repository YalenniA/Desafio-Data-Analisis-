```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# Calcular ingresos
ingreso_tienda1 = tienda1["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar resultados
print(f"Ingreso total - Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total - Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total - Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total - Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categor√≠a

```python
# Asegurar que los nombres de columnas est√©n limpios
for df in [tienda1, tienda2, tienda3, tienda4]:
    df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("√°", "a").str.replace("√©", "e").str.replace("√≠", "i").str.replace("√≥", "o").str.replace("√∫", "u")

# Contar productos vendidos por categor√≠a
ventas_cat_t1 = tienda1["categoria_del_producto"].value_counts()
ventas_cat_t2 = tienda2["categoria_del_producto"].value_counts()
ventas_cat_t3 = tienda3["categoria_del_producto"].value_counts()
ventas_cat_t4 = tienda4["categoria_del_producto"].value_counts()

# Mostrar resultados
print("üìä Ventas por categor√≠a - Tienda 1")
print(ventas_cat_t1)
print("\nüìä Ventas por categor√≠a - Tienda 2")
print(ventas_cat_t2)
print("\nüìä Ventas por categor√≠a - Tienda 3")
print(ventas_cat_t3)
print("\nüìä Ventas por categor√≠a - Tienda 4")
print(ventas_cat_t4)
```

```python
import matplotlib.pyplot as plt

# Visualizar ventas por categor√≠a de cada tienda
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

ventas_cat_t1.plot(kind="bar", ax=axs[0, 0], title="Tienda 1")
ventas_cat_t2.plot(kind="bar", ax=axs[0, 1], title="Tienda 2")
ventas_cat_t3.plot(kind="bar", ax=axs[1, 0], title="Tienda 3")
ventas_cat_t4.plot(kind="bar", ax=axs[1, 1], title="Tienda 4")

plt.tight_layout()
plt.show()
```

# 3. Calificaci√≥n promedio de la tienda


```python
# Calificaci√≥n promedio por tienda
promedio_t1 = tienda1["calificacion"].mean()
promedio_t2 = tienda2["calificacion"].mean()
promedio_t3 = tienda3["calificacion"].mean()
promedio_t4 = tienda4["calificacion"].mean()

# Mostrar resultados
print(f"‚≠ê Calificaci√≥n promedio - Tienda 1: {promedio_t1:.2f}")
print(f"‚≠ê Calificaci√≥n promedio - Tienda 2: {promedio_t2:.2f}")
print(f"‚≠ê Calificaci√≥n promedio - Tienda 3: {promedio_t3:.2f}")
print(f"‚≠ê Calificaci√≥n promedio - Tienda 4: {promedio_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos para el gr√°fico
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
promedios = [promedio_t1, promedio_t2, promedio_t3, promedio_t4]

# Crear gr√°fico
plt.figure(figsize=(8,5))
plt.bar(tiendas, promedios, color='skyblue')
plt.title("Calificaci√≥n Promedio por Tienda")
plt.ylabel("Calificaci√≥n (0 a 5)")
plt.ylim(0, 5)
plt.show()
```

```python
import matplotlib.pyplot as plt

def graficar_geografia(df, nombre_tienda):
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(df["lon"], df["lat"],
                          c=df["calificacion"],
                          cmap="viridis",
                          alpha=0.6)
    plt.colorbar(scatter, label="Calificaci√≥n")
    plt.title(f"Ubicaci√≥n de Ventas y Calificaciones - {nombre_tienda}")
    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.grid(True)
    plt.show()

graficar_geografia(tienda1, "Tienda 1")
graficar_geografia(tienda2, "Tienda 2")
graficar_geografia(tienda3, "Tienda 3")
graficar_geografia(tienda4, "Tienda 4")
```

```python
# Crear grupos de coordenadas redondeadas
tienda1["region"] = tienda1[["lat", "lon"]].round(1).astype(str).agg(" | ".join, axis=1)

# Promedios por regi√≥n
resumen_geo = tienda1.groupby("region").agg({
    "precio": "mean",
    "calificacion": "mean",
    "producto": "count"
}).rename(columns={"producto": "ventas"})

resumen_geo = resumen_geo.sort_values(by="ventas", ascending=False)
print(resumen_geo.head())
```
